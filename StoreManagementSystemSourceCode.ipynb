```python
#First of all we need to import mysql.connector package in our python script.
import mysql.connector  

#Next we need to establish connection to a MySQL database using connect() function of mysql.connector package.
#The connect() function of mysql.connector establishes connection to a MySQL database and requires four parameters,
#which are:
#<connection_object>=mysql.connector.connect (host=<host-name>,user=<username>,passwd=<password> [,database=<database>])
#- user is the username on MySQL
#- password is the password of the user
#- host-name is the database server hostname or IP address
#- database is optional which provides the databse name of a MySQL database
#The python identifier <connection_object> shall be used for executing queries and for other tasks on connected database.
connection_object=mysql.connector.connect(host="localhost",user="root",passwd="0000")

#We can check for successful connection using function is_connected() with connected object (which returns true, if connection is successful)
if connection_object.is_connected():
    print("CONNECTION TO THE MYSQL SERVER IS ESTABLISHED")
    print()    #These extra print statements are used to beautify the frontend of the software
    print()

#Now we create a database cursor instance by using cursor() function
# A Database Cursor is a special control structure that facilitates the row by row processing of records in the resultset.
#resultset is the set of records retreived as per query
cursor_object=connection_object.cursor()

#Once we have created a cursor, we can execute SQL queries using execute() function with the cursor object.
cursor_object.execute("CREATE DATABASE DILSANE_SINGH_CS_PROJECT")
print("DATABASE DILSANE_SINGH_CS_PROJECT CREATED")
print()
print()

cursor_object.execute("USE DILSANE_SINGH_CS_PROJECT")
cursor_object.execute("CREATE TABLE CUSTOMER_DETAILS(LOGIN_ID VARCHAR(50) NOT NULL PRIMARY KEY, PASSWORD VARCHAR(50) NOT NULL, CUSTOMER_NAME VARCHAR(50) NOT NULL, GENDER VARCHAR(20), MOB_NO BIGINT NOT NULL UNIQUE, ADDRESS VARCHAR(200))")
cursor_object.execute("CREATE TABLE PRODUCT_DETAILS(PRODUCT_ID VARCHAR(50) NOT NULL PRIMARY KEY, PRODUCT_NAME VARCHAR(50) NOT NULL, PRICE INTEGER NOT NULL)")
cursor_object.execute("CREATE TABLE ORDERS(LOGIN_ID VARCHAR(50) NOT NULL, PRODUCT_ID VARCHAR(50) NOT NULL , FOREIGN KEY (LOGIN_ID) REFERENCES CUSTOMER_DETAILS(LOGIN_ID) ON DELETE CASCADE ON UPDATE CASCADE, FOREIGN KEY (PRODUCT_ID) REFERENCES PRODUCT_DETAILS(PRODUCT_ID) ON DELETE CASCADE ON UPDATE CASCADE, PRODUCT_NAME VARCHAR(50) NOT NULL, QTY INTEGER NOT NULL ,  TOTAL_PRICE INTEGER NOT NULL, DATE DATE , TIME TIMESTAMP)")
cursor_object.execute("CREATE TABLE FEEDBACK(LOGIN_ID VARCHAR(50) NOT NULL, FOREIGN KEY (LOGIN_ID) REFERENCES CUSTOMER_DETAILS(LOGIN_ID) ON DELETE CASCADE ON UPDATE CASCADE, RATING INTEGER NOT NULL , DESCRIPTION VARCHAR(500) NOT NULL , DATE DATE , TIME TIMESTAMP)")
print("TABLES STRUCTURE CREATED")
print()
print()

H=["insert into product_details values ( 'TMK','TONNED MILK', 45)","insert into product_details values ( 'FMK','FULL CREAM MILK',55 )","insert into product_details values ('CCS' ,'CHOCOS',250 )","insert into product_details values ( 'BBD','BROWN BREAD', 45)","insert into product_details values ('WBD' ,'WHITE BREAD', 30)","insert into product_details values ( 'HST','HAND SANITIZER',40 )","insert into product_details values ( 'MSK','MASK',30 )","insert into product_details values ( 'GLV','GLOVES', 35)"]
for i in H:
    cursor_object.execute(i)
    connection_object.commit()
print("SOME RECORDS ARE ADDED SUCCESSFULY IN THE PRODUCT_DETAILS TABLE TO TEST THE PROGRAM")
print()
print()

H="insert into customer_details values ('DS','DS','DILSANE SINGH','MALE',8130057827,'12-A KV NMR, JNU')"
cursor_object.execute(H)
connection_object.commit()
print("YOU CAN USE THE LOGIN_ID=DS & PASSWORD=DS TO TEST THE PROGRAM")
print()
print()

print('DSOWNER IS THE LOGIN_ID OF STORE OWNER')
print('DSPASS IS THE PASSWORD OF STORE OWNER')
print()
print()
print("NOW YOU HAVE ENOUGH INFORMATION TO TEST THE PROGRAM")
```

```python
#First we shall import the packages/library/modules that are going to be used in the program.
#If the following packages are not installed  in your system them plesae first install them.
#To install tabulate you can run the following statement in an empty jupyter cell
#pip install tabulate
import mysql.connector
import datetime
from tabulate import tabulate

connection_object=mysql.connector.connect(host="localhost",user="root",passwd="0000",database="DILSANE_SINGH_CS_PROJECT")
if connection_object.is_connected():
    print("CONNECTION ESTABLISHED TO THE DATABASE DILSANE_SINGH_CS_PROJECT")
    print()
    print()
else:
    print("UNABLE TO ESTABLISH THE CONNECTION. PLEASE CONTACT THE STORE MANAGER TO HELP YOU")
    print()
    print()
cursor_object=connection_object.cursor()

#To make the software user friendly we are going to greet the user with the folllowing set of code lines
print("WELCOME TO THE DS GENERAL STORE")
print()
print()
print("PLEASE READ THE INSTRUCTIONS CAREFULLY")
print()
print()


#Now we should define the functions that we are going to use in the program.
#The most important reason to use functions is to make program handling easier as only a small part of the program is dealt with at a time, thereby avoiding ambiguity.
#Another reason to use functions is to reduce program size.
#Functions make a program more readable and understandable thereby making program management much easier.

def date_time_greet():
    x=datetime.datetime.now() #type(x) is datetime.datetime
    print(x.strftime("%A"))   # this prints current weekday
    print(x.strftime("%d"),x.strftime("%B"),x.strftime("%Y"))  # this prints current day of the month, month and year
    print("Time  :  ",x.strftime("%X")) # this prints current time
    print()
    print()

    
    
def see_a_customers_shopping_history(): 
    query="SELECT CUSTOMER_NAME, LOGIN_ID FROM CUSTOMER_details"
    cursor_object.execute(query)
    query_result=cursor_object.fetchall()
    D="Y"
    while D=="y" or D=="Y":
        print(tabulate(query_result,headers=['CUSTOMER_NAME','CUSTOMER_ID'],tablefmt='pretty'))
        print()
        print()
        LOGIN_ID_HS=input("ENTER THE CUSTOMER_ID OF THE CUSTOMER YOU WANT TO KNOW THE SHOPPING HISTORY OF  :  ")
        print()
        print()
        query2="SELECT * FROM ORDERS WHERE LOGIN_ID='{}'".format(LOGIN_ID_HS)
        cursor_object.execute(query2)
        query_result2=cursor_object.fetchall()
        print("CUSTOMER HISTORY")
        print()
        print(tabulate(query_result2,headers=['LOGIN_ID','PRODUCT_ID','PRODUCT_NAME','QTY','TOTAL_PRICE','DATE','TIME'],tablefmt="pretty"))
        D=input("DO YOU WANT TO SEE HISTORY OF ANY ELSE CUSTOMER? Y/N         :  ")
        print()
        print()
        
        
        
def delete_a_product():
    query="SELECT * FROM PRODUCT_DETAILS;"
    cursor_object.execute(query)
    query_result=cursor_object.fetchall()
    D="Y"
    while D=="y" or D=="Y":
        print(tabulate(query_result,headers=['PRODUCT_ID','PRODUCT_NAME','PRICE'],tablefmt='pretty'))
        print()
        print()
        PRODUCT_ID=input("ENTER THE PRODUCT ID OF THE PRODUCT YOU WANT TO DELETE     :  ")
        query="DELETE FROM PRODUCT_DETAILS WHERE PRODUCT_ID='{}'".format(PRODUCT_ID)
        cursor_object.execute(query)
        connection_object.commit()
        print("PRODUCT REMOVED SUCCESSFULLY")
        D=input("DO YOU WISH TO DELETE ANY OTHER PRODUCT? Y/N         :  ")
        print()
        print()
        
        
        
def add_a_new_product():
    D="Y"
    while D=="y" or D=="Y":
        PRODUCT_ID=input("ENTER PRODUCT_ID         :  ")
        PRODUCT_NAME=input("ENTER PRODUCT_NAME       :  ")
        PRICE=int(input("ENTER PRICE              :  "))
        print()
        print("PRODUCT ADDED SUCCESSFULY")
        query="INSERT INTO PRODUCT_DETAILS VALUES('{}','{}',{})".format( PRODUCT_ID, PRODUCT_NAME, PRICE)
        cursor_object.execute(query)
        connection_object.commit()
        D=input("DO YOU WISH TO ADD MORE PRODUCTS FOR SALE? Y/N         :  ")
        print()
        print()

        
        
def greet_after_logging_in_as_owner():
    date_time_greet()
    print("HELLO DILSANE SINGH !!!")
    print()
    print()
    print("PRESS 1 TO ADD A NEW PRODUCT")
    print("PRESS 2 TO DELETE A PRODUCT")
    print("PRESS 3 TO SEE A CUSTOMERS HISTORY")
    print("PRESS 4 TO RETURN TO MAIN MENU")
    print()
    print()
    CHOICE_greet_after_logging_in_as_owner=int(input("PLEASE ENTER YOUR CHOICE      :  "))
    print()
    print()
    print()
    print()
    if CHOICE_greet_after_logging_in_as_owner==1:
        add_a_new_product()
    if CHOICE_greet_after_logging_in_as_owner==2:
        delete_a_product()
    if CHOICE_greet_after_logging_in_as_owner==3:
        see_a_customers_shopping_history()
        
        
        
def login_as_owner():
    LOGIN_ID_OWNER=input("ENTER YOUR LOGIN_ID           :  ")
    if LOGIN_ID_OWNER=="DSOWNER":
        PASSWORD_OWNER=input("ENTER YOUR PASSWORD           :  ")
        if PASSWORD_OWNER=="DSPASS":
            print("LOGIN SUCCESSFUL")
            print()
            greet_after_logging_in_as_owner()
        else:
            print()
            print()
            print()
            print()
            print("INCORRECT PASSWORD, BUT CORRECT LOGIN_ID")
            print("PLEASE TRY AGAIN")
    else:
        print("INVALID LOGIN_ID")
        
        
        
def history(LOGIN_ID):
    query2="SELECT * FROM ORDERS WHERE LOGIN_ID='{}'".format(LOGIN_ID)
    cursor_object.execute(query2)
    query_result2=cursor_object.fetchall()
    if query_result2==[]:
        print("YOU HAVE NOT BOUGHT ANYTHING FROM THE STORE YET")
        print()
        print("WHAT ARE YOU WAITING FOR?")
        print()
        print("HURRY UP!!! GIVE US A CHANCE TO SERVE YOU")
    else:
        print("FOLLOWING IS YOUR SHOPPING HISTORY   :")
        print()
        print(tabulate(query_result2,headers=['LOGIN_ID','PRODUCT_ID','PRODUCT_NAME','QTY','TOTAL_PRICE','DATE','TIME'],tablefmt='pretty'))

        
        
def complaint(LOGIN_ID):
    RATING=int(input("RATE YOUR DIFFICULTY ON A SCALE OF 1 TO 10     :  "))
    DESCRIPTION=input("PLEASE DESCRIBE YOUR PROBLEM/COMPLAINT         :  ")
    cursor_object.execute("select curdate(), now()")
    r=cursor_object.fetchall()
    x=r[0][0]
    y=r[0][1]
    query="INSERT INTO FEEDBACK VALUES('{}',{},'{}','{}','{}')".format(LOGIN_ID, RATING, DESCRIPTION, x, y)
    print("FEEDBACK/COMPLAINT REGISTERED SUCCESFULLY")
    print()
    print()
    cursor_object.execute(query)
    connection_object.commit()
    
    
    
def save_to_order(c,LOGIN_ID):
    cursor_object.execute("select curdate(), now()")
    r=cursor_object.fetchall()
    x=r[0][0]
    y=r[0][1]
    for i in c:
        PRODUCT_NAME=i[1]
        PRODUCT_ID=i[0]
        QTY=i[3]
        TOTAL_PRICE=i[2]*i[3]
        query="INSERT INTO ORDERS VALUES('{}','{}', '{}',{},{},'{}','{}')".format(LOGIN_ID, PRODUCT_ID, PRODUCT_NAME, QTY, TOTAL_PRICE, x, y)
        cursor_object.execute(query)
        connection_object.commit()
    print("ORDER PLACED SUCCESFULLY")
    print()
    print()   

    
    
def show_cart(cart,LOGIN_ID):
    print()
    print()
    print("FOLLOWING IS THE SUMMARY OF YOUR CART")
    print()
    c=[]
    for i in cart:
        query1="SELECT PRODUCT_NAME FROM PRODUCT_DETAILS WHERE PRODUCT_ID='{}'".format(i[0])
        cursor_object.execute(query1)
        query_result=cursor_object.fetchall()
        D=query_result[0][0]
        query2="SELECT PRICE FROM PRODUCT_DETAILS WHERE PRODUCT_ID='{}'".format(i[0])
        cursor_object.execute(query2)
        query_result=cursor_object.fetchall()
        E=query_result[0][0]
        c.append([i[0],D,E,i[1]])
    print(tabulate(c,headers=['PRODUCT_ID','PRODUCT_NAME','PRICE',"QTY"],tablefmt='pretty'))
    a=0
    for i in c:
        a=a+(i[2]*i[3])
    print("TOTAL AMOUNT TO BE PAYED =",a,"Rs only/- ")
    print()
    print()
    save_to_order(c,LOGIN_ID)
    
    
    
def add_to_cart(LOGIN_ID):
    from tabulate import tabulate
    query="select * from product_details;"
    cursor_object.execute(query)
    query_result=cursor_object.fetchall()
    cart=list()
    D="y"
    while D=="y" or D=="Y":
        print(tabulate(query_result,headers=['PRODUCT_ID','PRODUCT_NAME','PRICE'],tablefmt='pretty'))
        print()
        print()
        b=input("ENTER PRODUCT_ID OF THE ITEM YOU WANT TO BUY       :  ")
        c=int(input("ENTER THE QUANTITY                                 :  "))
        print()
        cart.append([b,c])
        D=input("DO YOU WISH TO ADD MORE ITEMS TO CART? Y/N         :  ")
    show_cart(cart,LOGIN_ID)



def greet_after_logging_in(LOGIN_ID):
    query="SELECT CUSTOMER_NAME FROM CUSTOMER_DETAILS WHERE LOGIN_ID='{}'".format(LOGIN_ID)
    cursor_object.execute(query)
    query_result=cursor_object.fetchall()
    date_time_greet()
    print("HELLO ",query_result[0][0])
    print()
    print()
    print("PRESS 1 TO START SHOPPING")
    print("PRESS 2 TO REPORT A COMPLAINT/PROBLEM")
    print("PRESS 3 TO SEE YOUR SHOPPING HISTORY")
    print("PRESS 4 TO RETURN TO MAIN MENU")
    print()
    print()
    CHOICE_greet_after_logging_in=int(input("PLEASE ENTER YOUR CHOICE      :  "))
    print()
    print()
    print()
    print()
    if CHOICE_greet_after_logging_in==1:
        add_to_cart(LOGIN_ID)
    elif CHOICE_greet_after_logging_in==2:
        complaint(LOGIN_ID)
    elif CHOICE_greet_after_logging_in==3:
        history(LOGIN_ID)
       
    
    
def login_into_existing_account():
    LOGIN_ID=input("ENTER YOUR LOGIN_ID           :  ")
    query="SELECT * FROM CUSTOMER_DETAILS WHERE LOGIN_ID='{}'".format(LOGIN_ID)
    cursor_object.execute(query)
    query_result=cursor_object.fetchall()
    if query_result==[]:
        print("INVALID LOGIN_ID")
    else:
        PASSWORD=input("ENTER YOUR PASSWORD           :  ")
        query="SELECT PASSWORD FROM CUSTOMER_DETAILS WHERE LOGIN_ID='{}'".format(LOGIN_ID)
        cursor_object.execute(query)
        query_result=cursor_object.fetchall()
        if [(PASSWORD,)]==query_result:
            print("LOGIN SUCCESSFUL")
            print()
            print()
            greet_after_logging_in(LOGIN_ID)
        else:
            print()
            print()
            print("INCORRECT PASSWORD, BUT CORRECT LOGIN_ID")
            print("PLEASE TRY AGAIN")
    return(LOGIN_ID)



def create_new_account():
    CUSTOMER_NAME=input("ENTER YOUR NAME               :  ")
    MOB_NO=int(input("ENTER YOUR MOB_NO             :  "))
    GENDER=input("*ENTER YOUR GENDER            :  ")
    ADDRESS=input("*ENTER YOUR ADDRESS           :  ")
    LOGIN_ID=input("CREATE A LOGIN_ID             :  ")
    PASSWORD=input("CREATE A STRONG PASSWORD      :  ")
    query="INSERT INTO CUSTOMER_DETAILS VALUES('{}','{}','{}','{}',{},'{}')".format( LOGIN_ID, PASSWORD, CUSTOMER_NAME, GENDER, MOB_NO, ADDRESS)
    cursor_object.execute(query)
    connection_object.commit()
    print()
    print()
    print("ACCOUNT CREATED SUCCESSFULLY")
    print()
    print()

    
    
def greet():
    print()
    print("PRESS 1 TO CREATE A NEW ACCOUNT")
    print("PRESS 2 TO LOGIN INTO AN EXISTING CUSTOMER ACCOUNT")
    print("PRESS 3 TO LOGIN AS A STAFF")
    print("PRESS 4 TO QUIT")
    print()
    print()
    CHOICE_GREET=int(input("PLEASE ENTER YOUR CHOICE      :  "))
    print()
    print()
    return CHOICE_GREET



def menu():
    CHOICE_GREET=greet()
    if CHOICE_GREET==1:
        create_new_account()
        menu()
    elif CHOICE_GREET==2:
        login_into_existing_account()
        menu()
    elif CHOICE_GREET==3:
        login_as_owner()
        menu()
    else:
        print("Thankyou! Please visit our services!")

        
        
menu()
```
